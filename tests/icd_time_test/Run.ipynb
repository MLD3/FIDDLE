## Test: levels = [0]

```python
!mkdir -p output-0
```

```python
! PYTHONPATH="$PYTHONPATH:../../" \
python -m FIDDLE.run \
    --data_fname='./input/data.csv' \
    --population_fname='./input/pop.csv' \
    --config_fname='./input/config-0.yaml' \
    --output_dir='./output-0/' \
    --T=4 --dt=1.0 \
    --theta_1=0.001 --theta_2=0.001 --theta_freq=1 \
    --stats_functions 'min' 'max' 'mean'
```

```python
import numpy as np
import pandas as pd
import json
import sparse

X = sparse.load_npz('output-0/X_all.npz')
X_names = json.load(open('output-0/X_all.feature_names.json', 'r'))
X_index = pd.read_csv('output-0/X.ID,t_range.csv').set_index(['ID', 't_range'])
df_X = pd.DataFrame(X.todense().reshape(X.shape[0]*X.shape[1], X.shape[2]), columns=X_names, index=X_index.index)

display(df_X)
```
